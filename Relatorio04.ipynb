## Relações e Frequências

```python
import pandas as pd
dados = pd.read_csv("dados/alugiel_residencial.csv", sep=";")
dados
```

```python
# Selecione somente os imóveis classificados como apartamento

selecao = dados['Tipo'] == 'Apartamento'
n1 = dados[selecao].shape[0]
n1
```

```python
# Selecione somente os imóveis classificados como casa, casa de condomínio e casa de vila

selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao].shape[0]
n2
```

```python
# Selecione os imoveis com área entre 60 e 100 metros quadrados, incluindo os limites

selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n3 = dados[selecao].shape[0]
n3
```

```python
# Selecione os imoveis que tenham pelo menos 4 quartos, e aluguel menor que R$2.000,00

selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[selecao].shape[0]
n4
```

```python
print("Nº de imóveis classificados como apartamento: {}".format(n1))
print("Nº de imóveis classificados como casa, casa de condomínio e casa de vila: {}".format(n2))
print("Nº de imóveis com área entre 60 e 100 metros quadrados: {}".format(n3))
print("Nº de imóveis que tem pelo menos 4 quartos, e aluguel menor que R$2.000,00: {}".format(n4))
```
